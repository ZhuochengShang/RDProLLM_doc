```python
# POSIX classpath (macOS / Linux)
COMPILE_CLASSPATH = ":".join(DEPS)

# Spark --jars expects comma-separated
RUNTIME_JARS = ",".join(DEPS)

print("Compile classpath:\n", COMPILE_CLASSPATH)
print("\nRuntime jars:\n", RUNTIME_JARS)
```

```python
from pathlib import Path
import subprocess, shlex

RUN_DIR = Path("./runs/ndvi_eval")
RUN_DIR.mkdir(parents=True, exist_ok=True)

# 1) Clean (remove NOTES)
SRC_SCALA = Path("/Users/clockorangezoe/Documents/phd_projects/code/geoAI/RDProLLMagent/notebook/runs/manual_oracle/Job.manual.scala")
src = SRC_SCALA.read_text(encoding="utf-8")
clean = src.split("\nNOTES\n", 1)[0].rstrip() + "\n"

CLEAN_SCALA = RUN_DIR / "Job.cleaned.scala"
CLEAN_SCALA.write_text(clean, encoding="utf-8")
print("Wrote cleaned:", CLEAN_SCALA.resolve())

# 2) Make a combined Scala script (NO :load)
COMBINED = RUN_DIR / "combined.scala"
COMBINED.write_text(
    CLEAN_SCALA.read_text(encoding="utf-8"),
    encoding="utf-8"
)
print("Wrote combined:", COMBINED.resolve())

# 3) Beast jars
BEAST_HOME = Path("/Users/clockorangezoe/Documents/phd_projects/code/geoAI/RDProLLMagent/beast-0.10.2-SNAPSHOT-bin/beast-0.10.2-SNAPSHOT")
beast_jars = sorted([str(p) for p in (BEAST_HOME / "lib").glob("*.jar")])
RUNTIME_JARS = ",".join(beast_jars)

# 4) Run spark-shell non-interactively
SPARK_SHELL = "/Users/clockorangezoe/miniconda3/envs/geo_llm_spark/bin/spark-shell"
LOG_FILE = RUN_DIR / "spark_shell.log"

cmd = [
    SPARK_SHELL,
    "--master", "local[*]",
    "--jars", RUNTIME_JARS,
    "-i", str(COMBINED),   # âœ… run combined.scala, not run.scala
]

print("Running:\n", " ".join(map(shlex.quote, cmd)))
p = subprocess.run(cmd, text=True, capture_output=True)

LOG_FILE.write_text(p.stdout + "\n\n--- STDERR ---\n" + p.stderr, encoding="utf-8")

ok = (p.returncode == 0) and ("error:" not in (p.stdout + p.stderr).lower())
print("Return code:", p.returncode, "OK:", ok)
print("Log:", LOG_FILE.resolve())
print("\n--- Last 60 lines ---\n")
print("\n".join((p.stdout + "\n" + p.stderr).splitlines()[-60:]))

ok
```
